```python
import pandas as pd 
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import lifelines as ll
from lifelines import KaplanMeierFitter
```

```python
data = lifelines.datasets.load_dd()
```

```python
data
```

```python
kmf = KaplanMeierFitter()
T = data['duration']
C = data['observed']
kmf.fit(T,C)
```

```python
%pylab inline
figsize(12,6)
```

```python
kmf.plot()
```

```python
data['regime'].unique()
```

```python
ax = subplot(111)

for r in data['regime'].unique():
    ix = data['regime'] == r
    kmf.fit(T.ix[ix],C.ix[ix],label=r)
    ax = kmf.plot(ax=ax)
```
