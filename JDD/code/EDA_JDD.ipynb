```python
login_train = t_login_train.copy()
trade_train = t_trade_train.copy()
trade_train.rename(columns={'time':'trade_time'},inplace=True)
login_train.rename(columns={'time':'login_time'},inplace=True)
trade_train['trade_time'] = pd.to_datetime(trade_train['trade_time'])
trade_train.sort_values('trade_time',ascending=True)
trade_train.loc[:,'trade_time_diff'] = trade_train.groupby('id')['trade_time'].diff(1)
trade_train['trade_time_diff'] = trade_train['trade_time_diff'].map(lambda x: x.total_seconds())
login_trade_train = pd.merge(trade_train[['id','trade_time','trade_time_diff','is_risk']],login_train, on='id',how='left')
login_trade_train['is_risk'] = login_trade_train['is_risk'].astype('int')
# # login_trade_train['log_from'] = login_trade_train['log_from'].astype('int')
# login_trade_train['city'] = login_trade_train['city'].astype('int')
# login_trade_train['result'] = login_trade_train['result'].astype('int')
# login_trade_train['type'] = login_trade_train['type'].astype('int')
login_trade_train['trade_time'] = pd.to_datetime(login_trade_train['trade_time'])
login_trade_train['login_time'] = pd.to_datetime(login_trade_train['login_time'])
login_trade_train.loc[:,'login_trade_diff']= login_trade_train['trade_time'] - login_trade_train['login_time']
login_trade_train['login_trade_diff'] = login_trade_train['login_trade_diff'].map(lambda x: x.total_seconds())
login_trade_train
```
