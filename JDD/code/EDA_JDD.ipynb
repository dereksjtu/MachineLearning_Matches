### time proccessing
#### warm cold split
warm - trade log with history login info
    cold - trade log without history login info

```python
log_trade_tr = pd.read_csv('log_trade_tr.csv')
log_trade_tr['trade_time'] = pd.to_datetime(log_trade_tr['trade_time'])
log_trade_tr['login_time'] = pd.to_datetime(log_trade_tr['login_time'])
log_trade_tr.loc[:,'trade_log_diff'] = log_trade_tr['trade_time'] - log_trade_tr['login_time']
log_trade_tr['trade_log_diff'] = log_trade_tr['trade_log_diff'].map(lambda x: x.total_seconds())
log_trade_tr_warm = log_trade_tr[log_trade_tr['trade_log_diff'] >= 0]
warm = log_trade_tr_warm.groupby(['id','trade_time'],as_index=False)['login_time'].count()
trade_tr_cp = trade_tr.copy()
trade_tr_cp.rename(columns={'time':'trade_time'},inplace=True)
trade_tr_cp['trade_time'] = pd.to_datetime(trade_tr_cp['trade_time'])
trade_tr_cp = pd.merge(trade_tr_cp,warm,on=['id','trade_time'],how='left')
trade_tr_cold = trade_tr_cp[trade_tr_cp['login_time'].isnull()]
trade_tr_cold.sort_values(by=['id','trade_time'],inplace=True)
trade_tr_cold
```

### Time sliping

```python
# 1min
log_trade_tr_warm.loc[:,'trade_log_diff_1min'] = log_trade_tr_warm['trade_log_diff'] <= 60
log_trade_tr_warm['trade_log_diff_1min'] = log_trade_tr_warm['trade_log_diff_1min'].astype('int')
# 2min
log_trade_tr_warm.loc[:,'trade_log_diff_2min'] = log_trade_tr_warm['trade_log_diff'] <= 2*60
log_trade_tr_warm['trade_log_diff_2min'] = log_trade_tr_warm['trade_log_diff_1min'].astype('int')
#5min
log_trade_tr_warm.loc[:,'trade_log_diff_5min'] = log_trade_tr_warm['trade_log_diff'] < 5*60
log_trade_tr_warm['trade_log_diff_5min'] = log_trade_tr_warm['trade_log_diff_1min'].astype('int')
#10min
log_trade_tr_warm.loc[:,'trade_log_diff_10min'] = log_trade_tr_warm['trade_log_diff'] < 10*60
log_trade_tr_warm['trade_log_diff_10min'] = log_trade_tr_warm['trade_log_diff_1min'].astype('int')
#15min
log_trade_tr_warm.loc[:,'trade_log_diff_15min'] = log_trade_tr_warm['trade_log_diff'] < 15*60
log_trade_tr_warm['trade_log_diff_15min'] = log_trade_tr_warm['trade_log_diff_1min'].astype('int')
# 30min
log_trade_tr_warm.loc[:,'trade_log_diff_30min'] = log_trade_tr_warm['trade_log_diff'] < 30*60
log_trade_tr_warm['trade_log_diff_30min'] = log_trade_tr_warm['trade_log_diff_1min'].astype('int')
#60min
log_trade_tr_warm.loc[:,'trade_log_diff_60min'] = log_trade_tr_warm['trade_log_diff'] < 60*60
log_trade_tr_warm['trade_log_diff_60min'] = log_trade_tr_warm['trade_log_diff_1min'].astype('int')
#2h
log_trade_tr_warm.loc[:,'trade_log_diff_120min'] = log_trade_tr_warm['trade_log_diff'] < 2*60*60
log_trade_tr_warm['trade_log_diff_120min'] = log_trade_tr_warm['trade_log_diff_1min'].astype('int')
#5h
log_trade_tr_warm.loc[:,'trade_log_diff_300min'] = log_trade_tr_warm['trade_log_diff'] < 5*60*60
log_trade_tr_warm['trade_log_diff_300min'] = log_trade_tr_warm['trade_log_diff_1min'].astype('int')
log_trade_tr_warm
```
