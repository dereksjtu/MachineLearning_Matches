```python
# -*- encoding:urf-8 -*-
import pandas as pd
from sklearn.metrics import mean_squared_error
import numpy as np
from scipy.stats import mode
import warnings
warnings.filterwarnings('ignore')

train = pd.read_csv(u'训练集.csv',encoding='gbk')
print('训练集总数',train.shape)
print('字段',train.columns)
sub = pd.read_csv(u'用户提交.csv',encoding='gbk')
sub[u'销售日期_D'] = sub[u'日期'].map(lambda x:str(x)[-2:])
sub[u'销售日期_D'] = sub[u'销售日期_D'] .astype(int)
result = sub.copy()
sub
```

```python
sub_train = sub[[u'编码',u'销售日期_D',u'销量']]

train.loc[:,u'销售日期_M'] = train[u'销售日期'].map(lambda x:int(str(x)[5:6]))
train[u'销售日期_D'] = train[u'销售日期'].map(lambda x:int(str(x)[6:]))
type_list = list(train[u'商品类型'].unique())
train[u'是否促销'] = train[u'是否促销'].map(list(train[u'是否促销'].unique()).index)
train[u'商品类型'] = train[u'商品类型'].map(type_list.index)
train[u'单位'] = train[u'单位'].map(list(train[u'单位'].unique()).index)

train['is_buy'] = 1
train['is_buy'] = train['is_buy'].astype(int)

train_1 = train[train[u'销售日期_M'] == 1 ][[u'中类编码',u'销售日期_D','is_buy']]
train_1_count = train_1.groupby([u'销售日期_D',u'中类编码'],as_index=False)['is_buy'].sum()
train_1_count.rename(columns = {'is_buy':'is_buy_1'},inplace=True)
train_2 = train[train[u'销售日期_M'] == 2 ][[u'中类编码',u'销售日期_D','is_buy']]
train_2_count = train_2.groupby([u'销售日期_D',u'中类编码'],as_index=False)['is_buy'].sum()
train_2_count.rename(columns = {'is_buy':'is_buy_2'},inplace=True)

train_3 = train[train[u'销售日期_M'] == 3 ][[u'中类编码',u'销售日期_D','is_buy']]
train_3_count = train_3.groupby([u'销售日期_D',u'中类编码'],as_index=False)['is_buy'].sum()
train_3_count.rename(columns = {'is_buy':'is_buy_3'},inplace=True)

train_4 = train[train[u'销售日期_M'] == 4 ][[u'中类编码',u'销售日期_D','is_buy']]
train_4_count = train_4.groupby([u'销售日期_D',u'中类编码'],as_index=False)['is_buy'].sum()
train_4_count.rename(columns = {'is_buy':'is_buy_4'},inplace=True)
```

```python
sub_train.rename(columns={u'编码':u'中类编码'},inplace=True)
train_median_class = pd.merge(sub_train,train_1_count,on=[u'销售日期_D',u'中类编码'],how='left')
train_median_class = pd.merge(train_median_class,train_2_count,on=[u'销售日期_D',u'中类编码'],how='left')
train_median_class = pd.merge(train_median_class,train_3_count,on=[u'销售日期_D',u'中类编码'],how='left')
train_median_class = pd.merge(train_median_class,train_4_count,on=[u'销售日期_D',u'中类编码'],how='left')
```
