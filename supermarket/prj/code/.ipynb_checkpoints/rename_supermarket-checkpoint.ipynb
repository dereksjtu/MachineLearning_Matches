```python
# -*- coding:urf-8 -*-
# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Handle table like and matrices
import pandas as pd
import numpy as np

# Modeling Helper
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns
import missingno as msno

# Configure visualization
%matplotlib inline
mpl.style.use('ggplot')
sns.set_style('white')
pylab.rcParams['figure.figsize'] = 10,8

train_path = '../input/train.csv'
test_path = '../input/test.csv'
hol_path = '../input/holiday.csv'
train_date_path = '../input/train_date.csv'
cache_path = '../input/cache/'
output_path = '../output/'
train_new_path = 'train_new_total.csv'
train = pd.read_csv(train_path,encoding='gbk')
train_new = pd.read_csv(train_new_path)
test = pd.read_csv(test_path)
```

```python
train
```

```python
def timeHandle(s):
    s = str(s)
    s = [s[:4],s[4:6],s[6:]]
    return '-'.join(s)
```

```python
train.loc[:,'saleCount'] = 1
train_for_avg_o = train.groupby(['Class','SaleDate'],as_index=False)['saleCount'].sum()
train_for_avg_o.SaleDate = train_for_avg_o.SaleDate.map(lambda x: timeHandle(x))
train_for_avg_o['SaleDate'] = pd.to_datetime(train_for_avg_o['SaleDate'])
train_for_avg_o.loc[:,'dayOfWeek'] = train_for_avg_o.SaleDate.dt.dayofweek
train_for_avg_o.loc[:,'parClass'] = train_for_avg_o['Class'].map(lambda x: int(str(x)[:2]))
train_for_avg_o.to_csv('train_for_avg_o.csv')
```

```python
train_for_avg_b = train_for_avg_o[['parClass','SaleDate','saleCount']]
# train_for_avg_b = train_for_avg_o.groupby(['parClass','dayOfWeek'],as_index=False)['saleCount'].mean()
train_for_avg_b = train_for_avg_b.groupby(['parClass','SaleDate'],as_index=False)['saleCount'].sum()
train_for_avg_b.loc[:,'dayOfWeek'] = train_for_avg_b.SaleDate.dt.dayofweek
train_for_avg_b = train_for_avg_o.groupby(['parClass','dayOfWeek'],as_index=False)['saleCount'].mean()
train_for_avg_b['saleCount'] = train_for_avg_b['saleCount'].astype('int')
train_for_avg_b.rename(columns={'parClass':'Class'},inplace=True)
# train_for_avg_b[train_for_avg_b['parClass']==10]
# train_for_avg_b[[train_for_avg_b['parClass'] == 10] & [train_for_avg_b[train_for_avg_b['SaleDate'] == '2015-01-01']]]
# train_for_avg_b.to_csv('train_for_avg_b.csv',index=False)
train_for_avg_b
```

```python
train_for_avg_m = train_for_avg_o.groupby(['Class','dayOfWeek'],as_index=False)['saleCount'].mean()
train_for_avg_m['saleCount'] = train_for_avg_m['saleCount'].astype('int')
train_for_avg_m
```

```python
train_avg = pd.concat([train_for_avg_m,train_for_avg_b],axis=0)
train_avg
```

```python
test['SaleDate'] = test['SaleDate'].map(lambda x: timeHandle(x))
test['SaleDate'] = pd.to_datetime(test['SaleDate'])
test.loc[:,'dayOfWeek'] = test.SaleDate.dt.dayofweek
test
```

```python
result = pd.merge(test[['Class','SaleDate','dayOfWeek']],train_avg,on=['Class','dayOfWeek'],how='left')
result.fillna(0,inplace=True)
result['saleCount'] = result['saleCount'].astype('int')
del result['dayOfWeek']
test_o = pd.read_csv(test_path)
result['SaleDate'] = test_o['SaleDate']
result.rename(columns={'Class':u'编码','saleCount':u'销量','SaleDate':u'日期'},inplace=True)
result.to_csv('result_avg.csv',index=False,encoding='gbk')
result
```
