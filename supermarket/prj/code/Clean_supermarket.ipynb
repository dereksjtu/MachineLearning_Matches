```python
train_new = train_new[~train_new['Class'].isin(do_not_use_class)]
train_new.to_csv(train_valid_path + 'train_new_validate.csv')
```

```python
# saleCount mean
coord  = train_new.groupby('Class',as_index=False)['saleCount'].agg({'saleCount_mean':'mean'})
train_new = pd.merge(train_new, coord, on='Class', how = 'left')
coord  = train_new.groupby('Class',as_index=False)['saleCount'].agg({'saleCount_std':'std'})
train_new = pd.merge(train_new, coord, on='Class', how = 'left')
train_new
```

```python
train_new.loc[:,'saleCount_min'] = train_new['saleCount_mean'] - 2 * train_new['saleCount_std']
train_new.loc[:,'saleCount_max'] = train_new['saleCount_mean'] + 2 * train_new['saleCount_std']
train_new['saleCount_max'] = np.ceil(train_new['saleCount_max'])
train_new
```
